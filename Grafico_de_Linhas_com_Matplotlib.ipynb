Antes de qualquer coisa vamos ver para que serve um gráfico de linhas.

**Para que serve um gráfico de linhas?**

Um gráfico de linhas **serve muito bem quando se quer representar um dado ao longo do tempo**. 

No nosso exemplo, mostraremos a quantidade de torcedores em um estádio de futebol ao longo dos anos. Vamos usar os dados dos times: **Grêmio e Internacional**.

Primeiro vamos importar os pacotes, mencionar os nossos dados e definir as cores que vamos utilizar.

```python
# importando pacotes
import matplotlib.pyplot as plt
import numpy as np

# configurar as cores que usarei no plot
AZUL = "#003F5c" 
VERMELHO = "#FF0000"

# dados da qtd. de pessoas que entraram nos estádios 
ano = ["2015", "2016", "2017", "2018", "2019", "2020", "2021"]
inter = [21256, 20372, 19900, 23508, 24179, 24548, 12196]
gremio = [22749, 21572, 21649, 23857, 21557, 18812, 13458]
```

Em seguida, vamos plotar um gráfico de linhas simples. Abaixo, o significado de alguns atributos:

*   **figsize:** o tamanho da figura, largura e altura (eixo x e y).

*   **dpi:** define a resolução do gráfico (72 é uma resolução boa).

*   **range():** a quantidade colocada dentro dos parênteses. Refere-se a quantidade de anos, no caso, 7 (de 2015 a 2021). Que será o nosso eixo x.








```python
# definindo a figura
fig, ax = plt.subplots(figsize=(15,6), dpi=72)

# plots das linhas
ax.plot(range(7), inter, color=VERMELHO)
ax.plot(range(7), gremio, color=AZUL)

# exibir grafico
plt.show()
```

E agora vamos ajeitar um pouco. Vamos incluir mais alguns atributos, como:

*   **linewidth:** que refere-se a espessura da linha.

*   **ax.spines['right'].set_visible(False):** refere-se a remoção das bordas definidas de acordo com a necessidade (right - direita, top - acima).

*   **ax.set_ylim(10000, 25000):** é a definição do limite do eixo y, no caso, de 10000 a 25000.

E vamos incluir também a linha do tempo, incluindo os 7 anos, ajustando o eixo x (xticks), conforme abaixo:

```python
# definindo a figura
fig, ax = plt.subplots(figsize=(15,6), dpi=72)

# plots das linhas
ax.plot(range(7), inter, color=VERMELHO, linewidth=4)
ax.plot(range(7), gremio, color=AZUL, linewidth=4)

# remover as bordas da direita e superior
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

# mudar limite do eixo y
ax.set_ylim(10000, 25000)

# configurar xticks (colocar todos os rótulos, dos 7 anos)
ax.set_xticks(ticks=range(7))
ax.set_xticklabels(labels=ano)

# exibir grafico
plt.show()
```

Feito isso, vamos tirar os rótulos do eixo y e colocar os pontos na própria linha, para isso adicionaremos os códigos abaixo:

```python
# remover rótulos do eixo y (yticks)
ax.set_yticklabels(labels=[], visible=False)
ax.set_yticks(ticks=[])

# incluindo os pontos(scatter) nas linhas
ax.scatter(range(7), inter, s=90, color=VERMELHO, clip_on=False)
ax.scatter(range(7), gremio, s=90, color=AZUL, clip_on=False)
```


*   **s:** é o tamanho dos pontos.
*   **clip_on=False:** Serve para evitar de cortar a figura.


Pode acontecer do último ponto não aparecer, e mencionando dessa forma não corremos esse risco.

Ficou assim:

```python
# definindo a figura
fig, ax = plt.subplots(figsize=(15,6), dpi=72)

# plots das linhas
ax.plot(range(7), inter, color=VERMELHO, linewidth=4)
ax.plot(range(7), gremio, color=AZUL, linewidth=4)

# remover as bordas da direita e superior
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

# mudar limite do eixo y
ax.set_ylim(10000, 25000)

# configurar xticks (colocar todos os rótulos, dos 7 anos)
ax.set_xticks(ticks=range(7))
ax.set_xticklabels(labels=ano)

# remover rótulos do eixo y (yticks)
ax.set_yticklabels(labels=[], visible=False)
ax.set_yticks(ticks=[])

# incluindo os pontos(scatter) nas linhas
ax.scatter(range(7), inter, s=90, color=VERMELHO, clip_on=False)
ax.scatter(range(7), gremio, s=90, color=AZUL, clip_on=False)

# exibir grafico
plt.show()
```

Agora, digamos que só os últimos anos sejam importantes, e decidimos colocar só os três últimos pontos. Para isso, adicionaremos **[-3]**, conforme abaixo:



```python
# incluindo os pontos(scatter) nas linhas
ax.scatter(range(7)[-3:], inter[-3:], s=90, color=VERMELHO, clip_on=False)
ax.scatter(range(7)[-3:], gremio[-3:], s=90, color=AZUL, clip_on=False)
```

Veja abaixo como ficou:

```python
# definindo a figura
fig, ax = plt.subplots(figsize=(15,6), dpi=72)

# plots das linhas
ax.plot(range(7), inter, color=VERMELHO, linewidth=4)
ax.plot(range(7), gremio, color=AZUL, linewidth=4)

# remover as bordas da direita e superior
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

# mudar limite do eixo y
ax.set_ylim(10000, 25000)

# configurar xticks (colocar todos os rótulos, dos 7 anos)
ax.set_xticks(ticks=range(7))
ax.set_xticklabels(labels=ano)

# remover rótulos do eixo y (yticks)
ax.set_yticklabels(labels=[], visible=False)
ax.set_yticks(ticks=[])

# incluindo os pontos(scatter) nas linhas
ax.scatter(range(7)[-3:], inter[-3:], s=90, color=VERMELHO, clip_on=False)
ax.scatter(range(7)[-3:], gremio[-3:], s=90, color=AZUL, clip_on=False)

# exibir grafico
plt.show()
```

Vamos colocar agora os rótulos acima de cada um dos pontos. 

Pra fazer isso, vamos utilizar a estrutura `for`, e chamar o `ax.annotate` do Matplotlib para adicionar os rótulos. 

Vamos ver a definição de cada parâmetro que será utilizado dentro da estrutura `for`:

*   **xy:** são as coordenadas. Para x, definimos `i = 0 + 4`, indicando a partir de qual número queremos mostrar os rótulos, ou seja, a partir do ponto 5, que refere-se ao ano 2019. Já o `j`, refere-se ao eixo y, que nos trará as quantidades, ou seja, os rótulos.

*   **xytext:** é a distância do nosso rótulo, ajustando a posição conforme a necessidade (para cima, para a direita).

*   **textcoords="offset poits":**  Define os rótulos junto aos pontos.

*   **color:** a cor.

*   **fontsize:** tamanho da fonte do rótulo.

*   **fontweight:** a espessura da fonte do rótulo.

E então, vamos ver como ficou:

```python
# definindo a figura
fig, ax = plt.subplots(figsize=(15, 6), dpi=72)

# plots das linhas
ax.plot(range(7), inter, color=VERMELHO, linewidth=4)
ax.plot(range(7), gremio, color=AZUL, linewidth=4)

# incluindo os 3 últimos pontos(scatter) nas linhas
ax.scatter(range(7)[-3:], inter[-3:], s=90, color=VERMELHO, clip_on=False)
ax.scatter(range(7)[-3:], gremio[-3:], s=90, color=AZUL, clip_on=False)

# rótulos (anotações) dos pontos 
for i, j in enumerate(inter[-3:]):
    ax.annotate(
        j,
        xy=(i+4, j),
        xytext=(5, 10),
        textcoords='offset points',
        color=VERMELHO,
        fontsize=12,
        fontweight='medium'
    )

for i, j in enumerate(gremio[-3:]):
    ax.annotate(
        j,
        xy=(i+4, j),
        xytext=(5, 10),
        textcoords='offset points',
        color=AZUL,
        fontsize=12,
        fontweight='medium'
    )

# remover as bordas da direita e superior
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

# mudar range do eixo y
ax.set_ylim(10000, 25000)

# remover rótulos do eixo y (yticks)
ax.set_yticklabels(labels=[], visible=False)
ax.set_yticks(ticks=[])

# configurar xticks (colocar todos os rótulos, dos 7 anos)
ax.set_xticks(ticks=range(7))
ax.set_xticklabels(labels=ano)

# exibir grafico
plt.show()
```

Para finalizar, ajustamos o título e o rodapé, e prontinho.

```python
CINZA = "#9E9E9E"
CINZA_TEXTO = "#515151"

# definindo a figura
fig, ax = plt.subplots(figsize=(15, 6), dpi=72)

# plots das linhas
ax.plot(range(7), inter, color=VERMELHO, linewidth=4)
ax.plot(range(7), gremio, color=AZUL, linewidth=4)

# plots dos scatter plots
ax.scatter(range(7)[-3:], inter[-3:], s=90, color=VERMELHO, clip_on=False)
ax.scatter(range(7)[-3:], gremio[-3:], s=90, color=AZUL, clip_on=False)

# anotações nos pontos de scatter
for i, j in enumerate(inter[-3:]):
    ax.annotate(
        j,
        xy=(i+4, j),
        xytext=(5, 10),
        textcoords='offset points',
        color=VERMELHO,
        fontsize=12,
        fontweight='medium'
    )

for i, j in enumerate(gremio[-3:]):
    ax.annotate(
        j,
        xy=(i+4, j),
        xytext=(5, 10),
        textcoords='offset points',
        color=AZUL,
        fontsize=12,
        fontweight='medium'
    )

# remover spines da direita e superior
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

# mudar range do eixo y
ax.set_ylim(10000, 25000)

# remover yticks e yticklabels
ax.set_yticklabels(labels=[], visible=False)
ax.set_yticks(ticks=[])

# configurar xticks (colocar tocos ticks de 12 meses)
ax.set_xticks(ticks=range(7))
ax.set_xticklabels(labels=ano)

# title
ax.text(-0.3, 28500,'Quantidade anual de torcedores no estádio', fontsize=24, color=CINZA_TEXTO, fontweight='bold')
ax.text(-0.3, 26500,'Quantidade anual de pagantes que assistiram os jogos \nno estádio dos times Inter e Grêmio', fontsize=15, color=CINZA)

# rodapé
ax.text(
    -0.3, 8000,
    "Fonte: globo.com",
    fontsize=12, color=CINZA
    )

# exibir grafico
plt.show()
```
